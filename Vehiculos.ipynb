<h1>2.Preprocesamiento de datos 

```python
# Imputación de valores faltantes de manera segura (sin inplace=True)
Datos['Engine'] = Datos['Engine'].fillna(0)
Datos['Max Power'] = Datos['Max Power'].fillna(0)
Datos['Max Torque'] = Datos['Max Torque'].fillna(0)
Datos['Drivetrain'] = Datos['Drivetrain'].fillna('Unknown')
Datos['Length'] = Datos['Length'].fillna(Datos['Length'].median())
Datos['Width'] = Datos['Width'].fillna(Datos['Width'].median())
Datos['Height'] = Datos['Height'].fillna(Datos['Height'].median())
Datos['Seating Capacity'] = Datos['Seating Capacity'].fillna(Datos['Seating Capacity'].mode()[0])
Datos['Fuel Tank Capacity'] = Datos['Fuel Tank Capacity'].fillna(Datos['Fuel Tank Capacity'].median())
# Confirmar que los valores faltantes se han corregido
print(Datos.isnull().sum())
```

<h1>3.Seleccion de caracteristicas 

```python
# Asegurarse de que solo se usen las columnas numéricas para la correlación
numeric_cols = Datos.select_dtypes(include=[np.number])  # Selecciona solo columnas numéricas

# Calcular la correlación con 'Price'
correlaciones = numeric_cols.corr()['Price'].abs().sort_values(ascending=False)

# Seleccionar características que tienen una correlación mayor a 0.1
features = correlaciones[correlaciones > 0.1].index.tolist()

# Remover 'Price' de las características seleccionadas, ya que es nuestra variable objetivo
if 'Price' in features:
    features.remove('Price')

# Mostrar las características seleccionadas
print(f'Características seleccionadas: {features}')
```

<h1>4.División del Dataset en Train y Test

```python
# División en variables predictoras (X) y objetivo (y)
X = Datos[features]
y = Datos['Price']

# División en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.22, random_state=0)
```

<h1>5.Entrenamiento del modelo

```python
# Entrenamiento del modelo
model = LinearRegression()
model.fit(X_train, y_train)
```

<h1>6.Evaluar el modelo 

```python
# Predicción en el conjunto de prueba
y_pred = model.predict(X_test)

# Métricas de evaluación
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'MSE: {mse}')
print(f'MAE: {mae}')
print(f'R2 Score: {r2}')
```

<h1>7.visualizar Resultados

```python
# Gráfico de comparación entre precios reales y predichos
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, label='Predicciones', color='blue')
plt.plot(y_test, y_test, color='red', label='Valores Reales')  # Línea de regresión perfecta
plt.xlabel('Precio Real')
plt.ylabel('Precio Predicho')
plt.title('Comparación entre Precio Real y Predicho')
plt.legend()
plt.show()
```
